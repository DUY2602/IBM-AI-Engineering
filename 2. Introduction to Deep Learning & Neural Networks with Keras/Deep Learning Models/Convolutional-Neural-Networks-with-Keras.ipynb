## Import Keras and Packages


```python
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Input
from keras.utils import to_categorical
```

When working with convolutional neural networks in particular, we will need additional packages.


```python
from keras.layers import Conv2D # to add convolutional layers
from keras.layers import MaxPooling2D # to add pooling layers
from keras.layers import Flatten # to flatten data for fully connected layers
```

## Convolutional Neural Network with One Set of Convolutional and Pooling Layers


```python
# import data
from keras.datasets import mnist

# load data
(X_train, y_train), (X_test, y_test) = mnist.load_data()

# reshape to be [samples][pixels][width][height]
X_train = X_train.reshape(X_train.shape[0], 28, 28, 1).astype('float32')
X_test = X_test.reshape(X_test.shape[0], 28, 28, 1).astype('float32')
```

Let's normalize the pixel values to be between 0 and 1


```python
X_train = X_train / 255 # normalize training data
X_test = X_test / 255 # normalize test data
```

Next, let's convert the target variable into binary categories


```python
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)

num_classes = y_test.shape[1] # number of categories
num_classes
```

Next, let's define a function that creates our model. Let's start with one set of convolutional and pooling layers.


```python
def convolutional_model():
    
    # create model
    model = Sequential()
    model.add(Input(shape=(28, 28, 1)))
    model.add(Conv2D(16, (5, 5), strides=(1, 1), activation='relu'))
    model.add(MaxPooling2D(pool_size=(2, 2), strides=(2, 2)))
    
    model.add(Flatten())
    model.add(Dense(100, activation='relu'))
    model.add(Dense(num_classes, activation='softmax'))
    
    # compile model
    model.compile(optimizer='adam', loss='categorical_crossentropy',  metrics=['accuracy'])
    return model
```

| Layer | Type | Output Shape | Details |
|-------|------|-------------|---------|
| Input | `Input` | (28, 28, 1) | Grayscale image 28×28 |
| Conv2D | `Conv2D` | (24, 24, 16) | 16 filters, 5×5 kernel, stride 1×1, ReLU |
| MaxPooling | `MaxPooling2D` | (12, 12, 16) | Pool 2×2, stride 2×2 |
| Flatten | `Flatten` | (2304,) | 12 × 12 × 16 = 2304 |
| Hidden | `Dense` | (100,) | ReLU |
| Output | `Dense` | (num_classes,) | Softmax |

- **Conv2D**: slides 16 filters of size 5×5 across the image → output 24×24 because `28 - 5 + 1 = 24`  
- **MaxPooling2D**: takes the max value in each 2×2 window → halves spatial dimensions from 24×24 to 12×12  
- **Flatten**: converts 3D tensor (12×12×16) into 1D vector (2304,) before the Dense layers  
- **Optimizer**: Adam · **Loss**: Categorical Crossentropy · **Metric**: Accuracy

Finally, let's call the function to create the model, and then let's train it and evaluate it.


```python
# build the model
model = convolutional_model()

# fit the model
model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=10, batch_size=200, verbose=2)

# evaluate the model
scores = model.evaluate(X_test, y_test, verbose=0)
print("Accuracy: {} \n Error: {}".format(scores[1], (100-scores[1]*100) / 100))
```

## Convolutional Neural Network with Two Sets of Convolutional and Pooling Layers


Let's redefine our convolutional model so that it has two convolutional and pooling layers instead of just one layer of each.


```python
def convolutional_model():
    
    # create model
    model = Sequential()
    model.add(Input(shape=(28, 28, 1)))
    
    model.add(Conv2D(16, (5, 5), activation='relu'))
    model.add(MaxPooling2D(pool_size=(2, 2), strides=(2, 2)))
    
    model.add(Conv2D(8, (2, 2), activation='relu'))
    model.add(MaxPooling2D(pool_size=(2, 2), strides=(2, 2)))
    
    model.add(Flatten())
    model.add(Dense(100, activation='relu'))
    model.add(Dense(num_classes, activation='softmax'))
    
    # Compile model
    model.compile(optimizer='adam', loss='categorical_crossentropy',  metrics=['accuracy'])
    return model
```

| Layer | Type | Output Shape | Details |
|-------|------|-------------|---------|
| Input | `Input` | (28, 28, 1) | Grayscale image 28×28 |
| Conv2D | `Conv2D` | (24, 24, 16) | 16 filters, 5×5 kernel, ReLU |
| MaxPooling | `MaxPooling2D` | (12, 12, 16) | Pool 2×2, stride 2×2 |
| Conv2D | `Conv2D` | (11, 11, 8) | 8 filters, 2×2 kernel, ReLU |
| MaxPooling | `MaxPooling2D` | (5, 5, 8) | Pool 2×2, stride 2×2 |
| Flatten | `Flatten` | (200,) | 5 × 5 × 8 = 200 |
| Hidden | `Dense` | (100,) | ReLU |
| Output | `Dense` | (num_classes,) | Softmax |

- **Block 1**: Conv2D(16, 5×5) extracts low-level features (edges, curves) → MaxPooling halves spatial size from 24×24 to 12×12  
- **Block 2**: Conv2D(8, 2×2) extracts higher-level features on the compressed map → MaxPooling reduces to 5×5  
- **Flatten**: converts 3D tensor (5×5×8) into 1D vector (200,) before the Dense layers  
- Deeper architecture → learns more **abstract** features compared to single conv block  
- **Optimizer**: Adam · **Loss**: Categorical Crossentropy · **Metric**: Accuracy

## CNN Architecture Comparison

| | **CNN v1** | **CNN v2** |
|--|------------|------------|
| Conv blocks | 1 | 2 |
| Conv layers | Conv2D(16, 5×5) | Conv2D(16, 5×5) → Conv2D(8, 2×2) |
| Pooling layers | 1× MaxPooling2D | 2× MaxPooling2D |
| Shape after conv | (12×12×16) | (5×5×8) |
| Flatten output | (2304,) | (200,) |
| Dense hidden | 100 | 100 |
| Output | num_classes, Softmax | num_classes, Softmax |
| Parameters | more | fewer (smaller feature map) |

### Key Differences

- **Depth**: v2 adds a second conv block (Conv2D + MaxPooling) → learns more abstract, hierarchical features
- **Feature map size**: v1 flattens to 2304 neurons; v2 compresses down to only 200 → significantly fewer parameters
- **Filter progression**: v2 follows the common pattern of **decreasing filters** (16 → 8) as spatial size shrinks
- **Trade-off**: v2 is more compact and faster to train, but the extra pooling may lose some spatial detail

Now, let's call the function to create our new convolutional neural network, and then let's train it and evaluate it.


```python
# build the model
model = convolutional_model()

# fit the model
model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=10, batch_size=200, verbose=2)

# evaluate the model
scores = model.evaluate(X_test, y_test, verbose=0)
print("Accuracy: {} \n Error: {}".format(scores[1], (100-scores[1]*100) / 100))
```
